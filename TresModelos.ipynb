```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
from sklearn.datasets import make_blobs
from sklearn import datasets
from sklearn.svm import SVC
%matplotlib inline
```

```python
dataset = make_blobs(n_samples=200,n_features=2,centers=2,cluster_std=4,random_state=2)
X = pd.DataFrame(dataset[0])
y = pd.Series(dataset[1])
```

```python
plt.figure(figsize=(10,10))
plt.scatter(X.values[:, 0], X.values[:, 1], c=y.values, cmap=plt.get_cmap('PiYG'),edgecolors='k', s=100)
plt.title('Muestras aleatorias'); plt.xlabel('x'); plt.ylabel('y');
```

```python
import scipy 
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import train_test_split
from sklearn import svm
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)  
```

```python
plt.figure()
plt.scatter(X_train.values[:, 0], X_train.values[:, 1], c=y_train.values, cmap=plt.get_cmap('PiYG'), edgecolors='k', s=80)
plt.title('Train'); plt.xlabel('x'); plt.ylabel('y')
plt.show()
```

```python
# Importamos la biblioteca
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression().fit(X=X_train, y=y_train)
```

```python
# Training classifiers

x_min, x_max = X.values[:, 0].min() - 1, X.values[:, 0].max() + 1
y_min, y_max = X.values[:, 1].min() - 1, X.values[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))


Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, alpha=0.4, cmap=plt.get_cmap('PiYG'))
plt.scatter(X.values[:, 0], X.values[:, 1], c=y.values, alpha=0.8, cmap=plt.get_cmap('PiYG'), edgecolors='k', s=100, marker='o')
   

plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier(criterion='gini', max_depth=4, min_samples_split=5,min_samples_leaf=5)
dtc.fit(X=X_train, y=y_train)
```

```python
x_min, x_max = X.values[:, 0].min() - 1, X.values[:, 0].max() + 1
y_min, y_max = X.values[:, 1].min() - 1, X.values[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))


Z = dtc.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, alpha=0.4, cmap=plt.get_cmap('PiYG'))
plt.scatter(X.values[:, 0], X.values[:, 1], c=y.values, alpha=0.8, cmap=plt.get_cmap('PiYG'), edgecolors='k', s=100, marker='o')
   

plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier

#Solo se utiliza el vecino mas cercano al punto considerado k=1 k=n_neighbors  
knc = KNeighborsClassifier(n_neighbors=1, weights='distance', algorithm='auto', metric='minkowski')            
knc.fit(X=X_train, y=y_train)
```

```python
x_min, x_max = X.values[:, 0].min() - 1, X.values[:, 0].max() + 1
y_min, y_max = X.values[:, 1].min() - 1, X.values[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))


Z = knc.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, alpha=0.4, cmap=plt.get_cmap('PiYG'))
plt.scatter(X.values[:, 0], X.values[:, 1], c=y.values, alpha=0.8, cmap=plt.get_cmap('PiYG'), edgecolors='k', s=100, marker='o')
   

plt.show()
```
